```python
from transformers import AutoTokenizer, AutoModel, AutoConfig
```

```python
model = AutoModel.from_pretrained("hfl/rbt3", output_hidden_states=True, output_attentions=True)
```

```python
model.save_pretrained("./rbt3")
```

```python
config = AutoConfig.from_pretrained("./rbt3")
config
```

```python
tokenizer = AutoTokenizer.from_pretrained("./rbt3")
print(tokenizer)
```

```python
from transformers import BertConfig
config = BertConfig.from_pretrained("./rbt3")
config
```

```python
sen = "弱小的我也有大梦想！"
inputs = tokenizer([sen], add_special_tokens=True, return_tensors="pt", max_length=16, padding="max_length", truncation=True)
inputs
```

```python
outputs = model(**inputs)
outputs
```

```python
from transformers import AutoModelForSequenceClassification
model = AutoModelForSequenceClassification.from_pretrained("./rbt3", output_hidden_states=True, output_attentions=True)
model
```

```python
outputs = model(**inputs)
outputs
```

```python
import torch
output = torch.argmax(outputs.logits, dim=-1).item()
output
```

```python
model.config.id2label
```

```python
from transformers import pipeline
pipe = pipeline("text-classification", model=model, tokenizer=tokenizer)
```
