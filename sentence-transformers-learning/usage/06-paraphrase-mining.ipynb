# Paraphrase Mining

官方文档：https://sbert.net/examples/sentence_transformer/applications/paraphrase-mining/README.html

从一组句子中自动发现语义相似的句子对（释义对）。

```python
from sentence_transformers import SentenceTransformer
from sentence_transformers.util import paraphrase_mining

model = SentenceTransformer("all-MiniLM-L6-v2")

sentences = [
    "The cat sits outside",
    "A man is playing guitar",
    "I love pasta",
    "The new movie is awesome",
    "The cat plays in the garden",
    "A woman watches TV",
    "The new movie is so great",
    "Do you like pizza?",
]

# 自动发现所有释义对，按相似度降序排列
paraphrases = paraphrase_mining(model, sentences)

print("Top 5 释义对：")
for score, i, j in paraphrases[:5]:
    print(f"  [{score:.4f}] {sentences[i]} <-> {sentences[j]}")
```
