# Clustering

官方文档：https://sbert.net/examples/sentence_transformer/applications/clustering/README.html

将语义相似的文本自动分组。常用算法：K-Means、层次聚类。

```python
from sentence_transformers import SentenceTransformer
from sklearn.cluster import KMeans

model = SentenceTransformer("all-MiniLM-L6-v2")

sentences = [
    "Python是很好的编程语言", "Java适合企业开发", "Go语言性能很高",
    "今天天气真好", "明天会下雨", "气温骤降了",
    "深度学习很有趣", "神经网络能识别图像", "机器学习改变世界",
]

embeddings = model.encode(sentences)

# K-Means 聚类
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = kmeans.fit_predict(embeddings)

for cluster_id in range(3):
    print(f"\n--- Cluster {cluster_id} ---")
    for i, label in enumerate(labels):
        if label == cluster_id:
            print(f"  {sentences[i]}")
```

```python
# 层次聚类（Agglomerative）— 不需要预设簇数
from sklearn.cluster import AgglomerativeClustering

clustering = AgglomerativeClustering(
    n_clusters=None, distance_threshold=1.0
)
labels = clustering.fit_predict(embeddings)

print(f"自动发现 {len(set(labels))} 个簇：")
for cluster_id in set(labels):
    print(f"\n--- Cluster {cluster_id} ---")
    for i, label in enumerate(labels):
        if label == cluster_id:
            print(f"  {sentences[i]}")
```
